```python
# Import required libraries
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.datasets import make_blobs

# Step 1: Generate synthetic dataset
X, _ = make_blobs(n_samples=100, centers=3, cluster_std=1.0, random_state=42)

# Step 2: Compute the linkage matrix using Ward's method
linkage_matrix = sch.linkage(X, method='ward')

# Step 3: Plot dendrogram with annotations
plt.figure(figsize=(12, 6))
dendro = sch.dendrogram(linkage_matrix)

# Step 4: Add threshold lines to help determine number of clusters
thresholds = [7, 10, 15]  # Example thresholds to try
for thresh in thresholds:
    plt.axhline(y=thresh, color='r', linestyle='--', label=f'Threshold = {thresh}')

plt.title('Custom Dendrogram with Threshold Annotations')
plt.xlabel('Sample Index')
plt.ylabel('Distance')

# Avoid duplicate labels in legend
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

plt.show()
```
