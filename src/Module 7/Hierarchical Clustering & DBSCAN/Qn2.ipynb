```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# --------------------------
# Task 1: Visualizing Agglomerative Clustering (1D data)
# --------------------------

data_1d = np.array([1, 2, 5, 10, 15, 30]).reshape(-1, 1)

# Using 'complete' linkage as example
Z = linkage(data_1d, method='complete')

plt.figure(figsize=(8, 4))
dendrogram(Z, labels=data_1d.flatten())
plt.title('Agglomerative Clustering Dendrogram (1D data)')
plt.xlabel('Data Points')
plt.ylabel('Distance')

# Draw threshold line to choose clusters (example threshold)
threshold = 10
plt.axhline(y=threshold, color='r', linestyle='--', label=f'Threshold = {threshold}')
plt.legend()
plt.show()

print(f"Choosing threshold = {threshold} suggests cutting dendrogram into clusters where distance < threshold.")

# --------------------------
# Task 2: Visualizing Divisive Clustering (2D data)
# --------------------------

data_2d = np.array([(1, 2), (3, 4), (5, 5), (8, 9), (12, 15)])

# Divisive clustering is not natively supported in scipy,
# so we simulate by taking the agglomerative dendrogram and interpreting it upside-down,
# or by recursive KMeans
```
