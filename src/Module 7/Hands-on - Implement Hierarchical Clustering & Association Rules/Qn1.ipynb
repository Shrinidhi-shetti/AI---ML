```python
# Task 1: Hierarchical Clustering with Simple 2D Data
# Step 1: Import the necessary libraries.
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.datasets import make_blobs
# Step 2: Generate and visualize a simple 2D dataset.
n_samples = 100
n_features = 2
n_centers = 3
random_state = 42

X_2d, _ = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_centers, random_state=random_state)

plt.figure(figsize=(8, 6))
plt.scatter(X_2d[:, 0], X_2d[:, 1], s=50)
plt.title("Simple 2D Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()

# Step 3: Perform Hierarchical Clustering and plot the dendrogram.
linked_2d = linkage(X_2d, method='ward')

plt.figure(figsize=(12, 7))
dendrogram(linked_2d,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title("Hierarchical Clustering Dendrogram for Simple 2D Data (Ward Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()

# Task 2: Hierarchical Clustering with Iris Dataset
# Step 1: Import the Iris dataset.
from sklearn.datasets import load_iris
iris = load_iris()
X_iris = iris.data
y_iris = iris.target
feature_names = iris.feature_names
# Step 2: Perform Hierarchical Clustering and visualize using a dendrogram.
linked_iris = linkage(X_iris, method='ward')
plt.figure(figsize=(15, 8))
dendrogram(linked_iris,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True,
           labels=y_iris) # Using true labels for better visualization (optional)
plt.title("Hierarchical Clustering Dendrogram for Iris Dataset (Ward Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()

# Step 3: Determine the number of clusters from the dendrogram and interpret.
# Interpretation: From the dendrogram, by drawing a horizontal line that cuts the longest vertical lines
# without intersecting any cluster, we can visually determine the number of clusters.
# For the Iris dataset, typically, you'd observe 3 distinct clusters, which aligns with the true labels.
# Let's cut the dendrogram to get 3 clusters.
num_clusters_iris = 3
clusters_iris = fcluster(linked_iris, num_clusters_iris, criterion='maxclust')
print(f"\nIris Dataset Cluster Assignments ({num_clusters_iris} clusters):")
# We can print the first few assignments or analyze them
for i in range(10): # Print first 10 assignments
    print(f"Sample {i}: True Label {y_iris[i]}, Predicted Cluster {clusters_iris[i]}")

# Task 3: Hierarchical Clustering with Random Data
# Step 1: Generate random data.
n_samples_random = 150
n_features_random = 5
random_data = np.random.rand(n_samples_random, n_features_random)

# Step 2: Perform Hierarchical Clustering.
# Step 3: Experiment with different linkage methods such as 'single', 'complete', and observe the differences.

# Single Linkage
linked_single = linkage(random_data, method='single')
plt.figure(figsize=(15, 7))
dendrogram(linked_single,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title("Hierarchical Clustering Dendrogram for Random Data (Single Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()

# Complete Linkage
linked_complete = linkage(random_data, method='complete')
plt.figure(figsize=(15, 7))
dendrogram(linked_complete,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title("Hierarchical Clustering Dendrogram for Random Data (Complete Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()

# Average Linkage (as another example)
linked_average = linkage(random_data, method='average')
plt.figure(figsize=(15, 7))
dendrogram(linked_average,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title("Hierarchical Clustering Dendrogram for Random Data (Average Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()

print("\nDifferences in Linkage Methods:")
print("  - Single Linkage: Tends to produce 'long' clusters, susceptible to noise, and can lead to 'chaining' where clusters are stretched out.")
print("  - Complete Linkage: Tends to produce more compact, spherical clusters. Less susceptible to noise but can be sensitive to outliers.")
print("  - Average Linkage: A compromise between single and complete linkage. Tends to produce more balanced clusters than single linkage and less sensitive to outliers than complete linkage.")
print("  - Ward Linkage (used in Task 1 & 2): Aims to minimize the variance within each cluster. Often produces well-balanced, spherical clusters, especially with Euclidean distances.")
```
