Polynomial Regression

Task 1: Apply polynomial regression to a dataset showing the relationship between age and income.

```python
# Write your code here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample dataset: Age vs Income
data = {
    'Age': [22, 25, 28, 35, 40, 45, 50, 60],
    'Income': [25000, 28000, 35000, 50000, 52000, 51000, 48000, 40000]
}
df = pd.DataFrame(data)

# Prepare features
X = df[['Age']]
y = df['Income']

# Transform to polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit polynomial regression
model = LinearRegression()
model.fit(X_poly, y)

# Predict for plotting
age_range = np.linspace(min(X.Age), max(X.Age), 100).reshape(-1, 1)
age_range_poly = poly.transform(age_range)
income_pred = model.predict(age_range_poly)

# Plot
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(age_range, income_pred, color='red', label='Polynomial Regression')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Polynomial Regression: Age vs Income')
plt.legend()
plt.grid(True)
plt.show()
```

Task 2: Explore and visualize a higher-degree polynomial fitting compared to a simple linear regression.

```python
# Write your code here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample dataset: Age vs Income
data = {
    'Age': [22, 25, 28, 35, 40, 45, 50, 60],
    'Income': [25000, 28000, 35000, 50000, 52000, 51000, 48000, 40000]
}
df = pd.DataFrame(data)
X = df[['Age']]
y = df['Income']

# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_linear = lin_reg.predict(X)

# Polynomial Regression (degree 4)
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)

# Predict for smooth curve
age_range = np.linspace(min(X.Age), max(X.Age), 200).reshape(-1, 1)
age_range_poly = poly.transform(age_range)
y_pred_poly = poly_reg.predict(age_range_poly)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred_linear, color='green', label='Linear Regression')
plt.plot(age_range, y_pred_poly, color='red', label='Polynomial Regression (Degree 4)')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Linear vs Higher-Degree Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()
```

Task 3: Dataset with cyclical patterns such as tides or circadian cycles.

```python
# Write your code here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulate a dataset with cyclical (sinusoidal) patterns — e.g., tide levels over time
days = np.linspace(0, 30, 300)  # 30 days sampled
tide_levels = 2 * np.sin(2 * np.pi * days / 12.4) + np.random.normal(0, 0.2, len(days))  # 12.4-hour tidal cycle

# Create DataFrame
df = pd.DataFrame({'Day': days, 'TideLevel': tide_levels})

# Plotting the cyclical pattern
plt.figure(figsize=(10, 6))
plt.plot(df['Day'], df['TideLevel'], label='Tide Level', color='blue')
plt.title('Simulated Tidal Cycles Over 30 Days')
plt.xlabel('Day')
plt.ylabel('Tide Level (meters)')
plt.grid(True)
plt.legend()
plt.show()
```
