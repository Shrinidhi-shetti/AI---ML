```python
# ----------------------------
# Title: Biological vs Artificial Neural Networks
# ----------------------------

import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches

# ----------------------------
# Task 1: Human Brain vs Digital Neural Network
# ----------------------------

print("\n=== Task 1: Human Brain vs Artificial Neural Network ===")

similarities = [
    "1. Both use interconnected units (neurons) to process information.",
    "2. Both adjust connections to improve performance (learning).",
    "3. Both are capable of pattern recognition and decision making."
]

differences = [
    "1. Biological neurons communicate with electrochemical signals; artificial neurons use mathematical functions.",
    "2. Learning in the brain is based on synaptic plasticity; in ANNs, it's through weight updates using algorithms like backpropagation.",
    "3. Biological networks are highly energy efficient and parallel; artificial networks require much more energy and hardware."
]

print("\nSimilarities:")
for item in similarities:
    print(item)

print("\nDifferences:")
for item in differences:
    print(item)

# ----------------------------
# Task 2: Signal Processing Flowcharts
# ----------------------------

print("\n=== Task 2: Signal Processing Flowcharts ===")

def draw_flowcharts():
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    
    # Biological Neuron
    ax[0].set_title("Biological Neuron")
    ax[0].text(0.5, 0.9, "Dendrites", ha='center', fontsize=12, bbox=dict(boxstyle="round", facecolor='lightblue'))
    ax[0].text(0.5, 0.6, "Cell Body (Soma)", ha='center', fontsize=12, bbox=dict(boxstyle="round", facecolor='lightgreen'))
    ax[0].text(0.5, 0.3, "Axon", ha='center', fontsize=12, bbox=dict(boxstyle="round", facecolor='salmon'))
    ax[0].text(0.5, 0.05, "Synapse", ha='center', fontsize=12, bbox=dict(boxstyle="round", facecolor='orange'))
    ax[0].set_xlim(0, 1)
    ax[0].set_ylim(0, 1)
    ax[0].axis('off')

    # Artificial Neuron
    ax[1].set_title("Artificial Neuron")
    ax[1].text(0.5, 0.9, "Inputs (x₁, x₂, ...)", ha='center', fontsize=12, bbox=dict(boxstyle="round", facecolor='lightblue'))
    ax[1].text(0.5, 0.6, "Weights + Bias", ha='center', fontsize=12, bbox=dict(boxstyle="round", facecolor='lightgreen'))
    ax[1].text(0.5, 0.3, "Activation Function", ha='center', fontsize=12, bbox=dict(boxstyle="round", facecolor='salmon'))
    ax[1].text(0.5, 0.05, "Output", ha='center', fontsize=12, bbox=dict(boxstyle="round", facecolor='orange'))
    ax[1].set_xlim(0, 1)
    ax[1].set_ylim(0, 1)
    ax[1].axis('off')

    plt.suptitle("Signal Processing Flowcharts")
    plt.show()

draw_flowcharts()

# ----------------------------
# Task 3: Learning Mechanisms Comparison
# ----------------------------

print("\n=== Task 3: Learning Mechanisms ===")

data = {
    "Aspect": ["Trigger for Change", "Mechanism", "Time Scale", "Adaptability", "Stability"],
    "Synaptic Plasticity (Biological)": [
        "Neurotransmitter activity",
        "Changes in synapse strength (LTP/LTD)",
        "Milliseconds to years",
        "High (context-aware)",
        "High (homeostasis)"
    ],
    "Weight Adjustment (Artificial)": [
        "Loss/error signal",
        "Gradient descent (backpropagation)",
        "Milliseconds to seconds",
        "Depends on training data",
        "Depends on regularization"
    ]
}

df = pd.DataFrame(data)
print(df.to_string(index=False))
```
