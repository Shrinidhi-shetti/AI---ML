```python
# ----------------------------
# Title: AI vs ML vs DL Demo
# ----------------------------

# ----------------------------
# Task 1: AI in Chess (Stockfish)
# ----------------------------

import chess
import chess.engine
import time

print("\n=== Task 1: AI in Chess (Stockfish) ===")
stockfish_path = "/path/to/stockfish"  # üîÅ Replace with your Stockfish binary path

try:
    board = chess.Board()
    engine = chess.engine.SimpleEngine.popen_uci(stockfish_path)

    for _ in range(5):  # Limit to 5 moves for brevity
        result = engine.play(board, chess.engine.Limit(time=0.1))
        board.push(result.move)
        print(board)
        print("="*30)
        time.sleep(0.5)

    engine.quit()
except FileNotFoundError:
    print("‚ùå Stockfish binary not found. Please update 'stockfish_path' with the correct path.")

# ----------------------------
# Task 2: ML - Linear Regression
# ----------------------------

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

print("\n=== Task 2: ML - Linear Regression ===")

# Sample dataset
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1.5, 3.7, 2.9, 6.2, 7.1])

# Train model
model = LinearRegression()
model.fit(X, y)

# Predictions
X_test = np.array([[6], [7]])
predictions = model.predict(X_test)

# Plotting
plt.figure()
plt.scatter(X, y, color='blue', label='Training Data')
plt.plot(X, model.predict(X), color='red', label='Best Fit Line')
plt.scatter(X_test, predictions, color='green', label='Predictions')
plt.legend()
plt.title("Linear Regression Example (ML)")
plt.xlabel("X")
plt.ylabel("y")
plt.show()

print(f"Prediction for X=6: {predictions[0]:.2f}")
print(f"Prediction for X=7: {predictions[1]:.2f}")

# ----------------------------
# Task 3: DL - CNN for Image Classification
# ----------------------------

import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

print("\n=== Task 3: DL - CNN Image Classification ===")

# Load and preprocess data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28, 28, 1).astype("float32") / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype("float32") / 255.0

# Define CNN model
cnn_model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
cnn_model.fit(x_train, y_train, epochs=3, validation_split=0.1, verbose=1)

# Evaluate
test_loss, test_acc = cnn_model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_acc:.2f}")

# Predict a few test images
predictions = cnn_model.predict(x_test[:5])

# Show predictions
for i in range(5):
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title(f"Predicted: {np.argmax(predictions[i])}, Actual: {y_test[i]}")
    plt.axis('off')
    plt.show(
```
